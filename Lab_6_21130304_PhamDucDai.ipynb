#Task 1.
Task 1. Compare the performance of selected classification algorithms including **Random forest**, **NaiveBayes**, and **SVM** with **FASHION** dataset based on **accuracy, precision, recall, f1** measures according to **without using selection feature** and **using selection feature**.



```python
# code
algo_array = {
    0: 'Random forest',
    1: 'Navie Bayes MultinomialNB',
    2: 'SVM linear',
    3: 'Decision tree',
    4: 'Logistic Regression',
    5: 'kNN k=5'
}


def getScore(algo, X_train, X_test, y_train, y_test):
  if(algo==0):
    clf = RandomForestClassifier();
  elif(algo==1):
    clf = MultinomialNB()
  elif(algo==2):
    clf = svm.SVC(kernel='linear')
  elif(algo==3):
    clf = DecisionTreeClassifier()
  elif(algo==4):
    clf = LogisticRegression(max_iter=50000)
  elif(algo==5):
    clf = KNeighborsClassifier();
  clf.fit(X_train, y_train.values.ravel())
  y_pred = clf.predict(X_test);
  accuracy = metrics.accuracy_score(y_test, y_pred)
  precision = metrics.precision_score(y_test, y_pred, average='macro')
  recall = metrics.recall_score(y_test, y_pred, average='macro')
  f1 = metrics.f1_score(y_test, y_pred, average='macro')
  algo_name = algo_array.get(algo)
  return [algo_name, accuracy, precision, recall, f1]
```

```python
#code
fashion_data = pd.read_csv('fashion_train.csv')

X = fashion_data.drop(columns='y')
y = fashion_data[['y']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# print table
print('not using selection feature')
table_not_using = PrettyTable(['Algorithms', 'Accuracy', 'Precision', 'Recall', 'F1'])
for i in range(0,3):
  table_not_using.add_row(getScore(i, X_train, X_test, y_train, y_test))
print(table_not_using)

new_X = SelectKBest(chi2, k=round(fashion_data.shape[1] / 2)).fit_transform(X, y)
X_train, X_test, y_train, y_test = train_test_split(new_X, y, test_size=0.2)
print('using selection feature')
table_using = PrettyTable(['Algorithms', 'Accuracy', 'Precision', 'Recall', 'F1'])
for i in range(0,3):
  table_using.add_row(getScore(i, X_train, X_test, y_train, y_test))
print(table_using)
```

#2.1. Apply StandardScaler() function to columns that contains numerical data ('age', 'balance', 'day', 'campaign', 'pdays', 'previous')

```python
def my_scaler(data, columns):
  scaler = StandardScaler()
  for i in columns:
    data[[i]] = scaler.fit_transform(data[[i]])
    data[[i]] += abs(data[[i]].min())
  return data
```

```python
#code
bank_data = pd.read_csv('bank.csv')
columns = ['age', 'balance', 'day', 'campaign', 'pdays', 'previous']
new_bank_data = my_scaler(bank_data, columns)
new_bank_data
```

#2.2. Apply Encode Categorical Value (OneHotEncoder) to transfrom categorical data to numerical data ('job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome')

```python
def my_label_encoder(data, columns):
  label_encoder = LabelEncoder()
  for i in columns:
    data[i] = label_encoder.fit_transform(data[i])
  return data
```

```python
#code
columns = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
new_bank_data = my_label_encoder(bank_data, columns)
new_bank_data
```
